```python
import pandas as pd
import numpy as np
import matplotlib
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
```

```python
data = pd.read_feather('../server/files/digits-toy.feather')
y_column="target"
xt, xs, yt, ys = train_test_split(
        data.drop(labels=[y_column], axis=1),
        data[y_column],
        test_size=0.3
)
```

```python
vt = VarianceThreshold(threshold=0.2)
vt.fit(xt)
pd.DataFrame(vt.transform(xt), columns=xt.columns[vt.get_support()])
```
